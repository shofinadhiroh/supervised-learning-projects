```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# bikin 0 sampe 10 sebanyak 100 titik?
X = np.linspace(0, 10, 100)
y = np.sin(X)
```

## MATLAB Style

```python
plt.plot(X, y)
```

## Matplotlib Style

```python
fig = plt.figure() # bikin canvas
ax = plt.axes()
ax.plot(X, y)
```

# Standard Example

## Matlab Style

```python
plt.figure(figsize=(8, 6))
plt.plot(X, y, "r--", linewidth=3, label="sinus")
plt.plot(X, np.cos(X), 'b-.', linewidth=3, label="cosinus")
plt.xlim(0, 15)
plt.ylim(-2, 2)
plt.title("Trigonometri")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

## Matplotlib Style

```python
fig = plt.figure(figsize=(8, 6))
ax = plt.axes()
ax.plot(X, y, "r--", linewidth=3, label="sinus")
ax.plot(X, np.cos(X), 'b-.', linewidth=3, label="cosinus")
ax.set(xlim=(0,15), ylim=(-2,2), xlabel="x", ylabel="y", title="Trigonometri")
ax.legend();
```

    'b' = warna biru, solid line
    'ro' = warna merah, marker bulat
    'g-' = warna hijau, solid line
    '--' = warna default, garis dashed
    'k^:' = warna hitam, marker segitiga, garis dotted
    'c+-' = warna cyan, marker +, solid line

# Subplots

## Matlab style

```python
plt.subplot(221)
plt.plot(X, y)

plt.subplot(224)
plt.plot(X, np.cos(X))
```

## Matplotlib style

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))

axes[0, 0].plot(X, y)
axes[1, 1].plot(X, np.cos(X))
```
