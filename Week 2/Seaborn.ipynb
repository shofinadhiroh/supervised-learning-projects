```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/linear.csv')
df.head()
```

```python
df1 = df.query('tipe == "data1"')
df2 = df.query('tipe == "data2"')
```

```python
plt.scatter(df1.X, df1.y, label='data1')
plt.scatter(df2.X, df2.y, label='data2')
plt.xlabel("x")
plt.ylabel("y")
plt.legend();
```
