# Mendapatkan semua image path

```python
from glob import glob
```

```python
glob("data/*/*.png")
```

```python
fname = glob("data/*/*.png")
fname[:10]
```

# X = Path -> Image -> Flatten

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

pipeline data engineering

```python
X = np.array([plt.imread(f).flatten() for f in fname])
X
```

```python
X.shape
```

# y = Path -> split -> ambil [1]

```python
y = np.array([int(f.split("\\")[1]) for f in fname])
y
```

```python
y.shape
```

# Join sebagai 1 dataframe lalu save as csv

```python
df = pd.DataFrame(X, columns=range(1,785))
df.head()
```

```python
df['label'] = y
df.head()
```

# Save to csv

```python
df.to_csv("metadata_exercise.csv", index=None)
```
